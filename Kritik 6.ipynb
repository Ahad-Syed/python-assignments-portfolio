```python
import numpy as np
import matplotlib.pyplot as plt

def density(mean, variance, x):
    std_dev = np.sqrt(variance)
    exponent = -((x - mean)**2) / (2 * variance)
    density = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(exponent)
    return density

# Values for x
x_values = np.linspace(-5, 5, 1000)

# Mean and variance values
mean_values = [0, 2, -1]
variance_values = [1, 0.5, 2]

# Plotting the density functions for different mean and variance values
plt.figure(figsize=(10, 6))

for mean, variance in zip(mean_values, variance_values):
    density_values = [density(mean, variance, x) for x in x_values]
    label = f'Mean={mean}, Variance={variance}'
    plt.plot(x_values, density_values, label=label)

plt.title('Normal Distribution Density Functions')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
