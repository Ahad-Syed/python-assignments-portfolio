```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point):
    # Function to calculate the derivative of f at a given point
    def deriv(f, base_point):
        return (f(base_point + 1e-5) - f(base_point)) / 1e-5

    # Initialize lists to store the coordinates for plotting
    x_coordinate = [initial_point]
    y_coordinate = [f(initial_point)]

    # Perform gradient descent iterations
    for _ in range(10000):
        # Calculate the gradient at the current point
        gradient = deriv(f, x_coordinate[-1])

        # Update the current point using the gradient and learning rate
        update = -learning_rate * gradient
        x_coordinate.append(x_coordinate[-1] + update)
        y_coordinate.append(f(x_coordinate[-1]))

    # Generate a range of x values for plotting the function
    plot_range = np.linspace(min(x_coordinate) - 0.5, max(x_coordinate) + 0.5, 10000)
    
    # Calculate the corresponding y values for the function plot
    function_range = [f(i) for i in plot_range]

    # Plot the function and the gradient descent path
    plt.plot(plot_range, function_range, label='Function')
    plt.plot(x_coordinate, y_coordinate, marker='o', color='r', label='Gradient Descent')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.title('Descent Optimization')
    plt.legend()
    plt.show()

    # Return the final optimized value of x
    return round(x_coordinate[-1], 3)
    
def f(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

        
learning_rate = 0.1
initial_point = -1.0
result = gradient_descent(f, learning_rate, initial_point)
print('Local Minima:', result)
```
